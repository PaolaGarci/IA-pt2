```python
import numpy as np
import pandas as pd

from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error

from SeqMetrics import RegressionMetrics
```

```python
df = pd.read_csv('/content/drive/MyDrive/IAConcentracion2/Copia de 2012_2019_PlatteRiverWeir_features_merged_all.csv')
df
```

```python
df.columns = ['SensorTime', 'CaptureTime', 'Filename', 'Agency', 'SiteNumber',
       'TimeZone', 'Stage', 'Discharge', 'CalcTimestamp', 'width',
       'height', 'exposure', 'fNumber', 'isoSpeed', 'shutterSpeed',
       'grayMean', 'graySigma', 'entropyMean', 'entropySigma', 'hMean',
       'hSigma', 'sMean', 'sSigma', 'vMean', 'vSigma', 'areaFeatCount',
       'grayMean 0', 'graySigma 0', 'entropyMean 0', 'entropySigma 0',
       'hMean 0', 'hSigma 0', 'sMean 0', 'sSigma 0', 'vMean 0',
       'vSigma 0', 'grayMean 1', 'graySigma 1', 'entropyMean 1',
       'entropySigma 1', 'hMean 1', 'hSigma 1', 'sMean 1', 'sSigma 1',
       'vMean 1', 'vSigma 1', 'WeirAngle', 'WeirPt1X', 'WeirPt1Y',
       'WeirPt2X', 'WeirPt2Y', 'WwRawLineMin', 'WwRawLineMax',
       'WwRawLineMean', 'WwRawLineSigma', 'WwCurveLineMin',
       'WwCurveLineMax', 'WwCurveLineMean', 'WwCurveLineSigma']

df = df.drop(['Filename', 'Agency', 'SiteNumber','TimeZone'],axis=1)
df
```

```python
df = df.drop(['SensorTime', 'CaptureTime',  'CalcTimestamp'],axis=1)
df
```

```python
mlp = MLPRegressor() #hidden_layer_sizes=(2,),activation='logistic',alpha=0.01
mlp.fit(df.drop(['Stage','Discharge'],axis=1),df[['Stage','Discharge']])

y_predMLP = mlp.predict(df.drop(['Stage','Discharge'],axis=1))
```

```python
y_predMLP = pd.DataFrame(y_predMLP,columns=['Stage','Discharge'])
y_predMLP
```

```python
erMLPS_S = RegressionMetrics(df['Stage'], y_predMLP['Stage'])
erMLP_D = RegressionMetrics(df['Discharge'], y_predMLP['Discharge'])
```

```python
svr_S = SVR()
svr_D = SVR()

svr_S.fit(df.drop(['Stage','Discharge'],axis=1),df['Stage'])
svr_D.fit(df.drop(['Stage','Discharge'],axis=1),df['Discharge'])

y_predSVR_S = svr_S.predict(df.drop(['Stage','Discharge'],axis=1))
y_predSVR_D = svr_D.predict(df.drop(['Stage','Discharge'],axis=1))
```

```python
y_predSVR = pd.DataFrame({'Stage':y_predSVR_S,'Discharge':y_predSVR_D})
y_predSVR
```

```python
erSVR_S = RegressionMetrics(df['Stage'], y_predSVR['Stage'])
erSVR_D = RegressionMetrics(df['Discharge'], y_predSVR['Discharge'])
```

```python
rfr = RandomForestRegressor()
rfr.fit(df.drop(['Stage','Discharge'],axis=1),df[['Stage','Discharge']])
y_predRFR = rfr.predict(df.drop(['Stage','Discharge'],axis=1))
```

```python
y_predRFR = pd.DataFrame(y_predRFR,columns=['Stage','Discharge'])
y_predRFR
```

```python
erRFR_S = RegressionMetrics(df['Stage'], y_predRFR['Stage'])
erRFR_D = RegressionMetrics(df['Discharge'], y_predRFR['Discharge'])
```

```python
def metricas(err):
  errores = {'R2':err.r2_score(),'MAE':err.mae(),'MSE':err.mse(),'RMSE':err.rmse(),'NRMSE':err.nrmse(),'PBIAS':err.pbias(),'NSE':err.nse()}
  return errores
```

```python
obj_errores = [erMLPS_S,erMLP_D,erSVR_D,erSVR_S,erRFR_D,erRFR_S]
errores = []

for error in obj_errores:
  errores.append(metricas(error))
```

```python
for modelo,error in zip(['MLP Stage','SVR Stage','RFR Stage','MLP Discharge','SVR Discharge','RFR Discharge'],errores):
  print(modelo+f":\n\tR2: {error['R2']} \n\tMAE: {error['MAE']}\n\tMSE: {error['MSE']}\n\tRMSE: {error['RMSE']} \n\tNRMSE: {error['NRMSE']} \n\tPBIAS: {error['PBIAS']} \n\tNSE: {error['NSE']}")
```
